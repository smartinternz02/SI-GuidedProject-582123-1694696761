```python
#Pie plot of the df(Head)
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(dfh['number of floors'],autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of number of floors')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#pie plot of df(tail)
plt.figure(figsize=(6, 6))  # Set the figure size
plt.pie(dfl['Price'],[0,0,0.1,0,0],labels=['t1','t2','t3','t4','t5'],autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Prices')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
#histogram plot of df(head)
plt.figure(figsize=(6, 6))
sns.histplot(dfh['Price'], kde=True)
plt.title('Histogram of Price')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#Barplot of df(head)
plt.figure(figsize=(6, 6))
plt.bar(dfh['living area'], dfh['Price'])
plt.title('Bar Chart of living area')
plt.xlabel('living area')
plt.ylabel('prices')
plt.show()
```

```python
#linegraph of df(head)
plt.figure(figsize=(6, 6))
plt.plot(dfh['living area'], dfh['Price'])
plt.title('Line Graph of living area')
plt.xlabel('living area')
plt.ylabel('prices')
plt.show()
```

```python
#Satterplot of df(head)
plt.figure(figsize=(6, 6))
plt.scatter(dfh['living area'], dfh['Price'])
plt.title('Scatter Plot of living area')
plt.xlabel('living area')
plt.ylabel('prices')
plt.show()
```

```python
null_values = df.isnull().sum()
null_values
```

```python
# Identify columns with missing values
missing_values = df.columns[df.isnull().any()]
missing_values
```

```python
# Calculate the median for each column with missing values and store it in a dictionary
median_values = {}
for column in missing_values:
    median = df[column].median()
    median_values[column] = median

median_values
df.fillna(median_values, inplace=True)
```
